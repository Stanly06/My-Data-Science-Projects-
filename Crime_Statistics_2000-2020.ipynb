Para detectar valores atípicos en la variable "Number of Crimes" podemos utilizar un gráfico de cajas y bigotes. Este tipo de gráfico muestra la distribución de los datos y nos permite visualizar si hay valores que se alejan mucho de la media y los valores típicos.

```python
import seaborn as sns

sns.boxplot(x=df['Number of Crimes'])
```

Se procede a identificar el valor atipico que se muestra en el grafico de cajas y bigotes, una forma efectiva es ir a la tabla de analisi estadistico y ver cual es el valor maximo que segun el grafico se aleja bastante con relacion a la media, el cual es 6203

```python
df = df[df['Number of Crimes'] != 6203]
```
